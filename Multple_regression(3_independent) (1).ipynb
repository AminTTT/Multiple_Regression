```python
df = pandas.read_csv("car.csv")
```

```python
X = df[['Weight', 'Volume', 'HP']]
y = df['CO2']
```

```python
train_X = X[2:33]
train_y = y[2:33]



test_X = X[33:]
test_y = y[33:]
print(test_X)
# توجه توجه : ترتیب ستون ها بین آنچه پایتون پرینت کرده و آنچه در ترتیب ستون ها در اکسل است ، متفاوت میباشد
```

```python
from sklearn import linear_model
```

```python
#  From the sklearn module we will use the LinearRegression() method to create a linear regression object.
# This object has a method called fit() that takes the independent and dependent values as parameters and...
# fills the regression object with data that describes the relationship:
#
```

```python
regr = linear_model.LinearRegression()
regr.fit(train_X, train_y)
```

```python
#  Now we have a regression object that are ready to predict CO2 values based on a car's weight , volume and HP:
```

```python
# پیش بینی داده های تست
predictedCO2 = regr.predict(test_X)
print(predictedCO2)
```

```python
# پیش بینی چند مقدار که به صورت دستی وارد میشوند
# دقت  دقت دقت: ترتیب ستون ها بین آنچه پایتون قرار داده و آنچه در اکسل است متفاوت است. به چند دستور بالاتر ...
# ... که تست ایکس را پرینت کردی توجه کن
predictedCO2 = regr.predict([[1390, 1600,271],[1405, 1600, 238]])
print(predictedCO2)
```

```python
#  In this case, we can ask for the coefficient value of weight against CO2,  volume against CO2 and  HP against CO2.
# یعنی تاثیر وزن خودرو روی تولید کربن دی اکسید چقدر است؟ و  تاثیر حجم خودرو و همینطور تاثیر قدرت خودرو چقدر است؟
# The answer(s) that we get tells us what would happen if we increase,or decrease,  one of the independent values.
```

```python
print(regr.coef_)
```
