**To produce and seperate train and test data**

```python

train_X = X[0:35]
train_y = y[0:35]



test_X = X[36:]
test_y = y[36:]
# actual=test_y

for i in test_y:
  test_y=i
  print(test_y)


```

***Using linearegression to describe relataionship between 2 Independent and one dependent values ***

```python
from sklearn import linear_model
# From the sklearn module we will use the LinearRegression() method to create a linear regression object.
# This object has a method called fit() that takes the independent and dependent values as parameters and fills the regression object with data that describes the relationship:
regr = linear_model.LinearRegression()
#for num in train_X
regr.fit(train_X, train_y)
predicted_Percent = regr.predict(test_X)
print(predicted_Percent)
print('Live=1 OR dead=0')
for num in predicted_Percent:
  if num >0.5:
    predicted_Percent=1
  else:
    predicted_Percent=0
  print(predicted_Percent)

```

```python
# These code line  tell us that if which one of Age and Period parameter of illness increase by 1 Unit , how much the percentage of health become increase or decrease.
print(regr.coef_)
```
